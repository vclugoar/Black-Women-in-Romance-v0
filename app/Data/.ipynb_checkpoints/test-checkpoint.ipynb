```python
import pandas as pd 

df = pd.read_csv('Books2.csv')
```

```python
df.columns
```

```python
df2 = df[['index', 'Book', 'Author', 'Image', 'Year', 'Description', 'Buy',  'h0', 'h1', 
       'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'h8', 'h9', 'h10', 'h11', 'h12',
       'h13', 'h14', 'h15', 'h16', 'h17', 'h18', 'h19', 'h20', 'h21', 'h22',
       'h23', 'h24', 'h25', 'h26', 'h27', 'h28', 'h29']]

df2
```

```python
melted_df = df2.melt(id_vars=['index', 'Book', 'Author', 'Image', 'Year', 'Description', 'Buy'], 
        var_name= 'Topics', 
        value_name="Value")

melted_df = melted_df[melted_df['Value'] > 0]

melted_df[melted_df['Book'] == 'Stepping to a New Day']
```

0 & 8 - action/crime
2 - Sex/Adult/Explicit
7 - Family/Business?
13 - Slavery 
17 & 23 & 24 - Tea/Royal
22- Lawyer/Media 
25 - Performance/Audience

```python
melted_df2 = melted_df[melted_df['Topics'].isin(['h0', 'h8', 'h2', 'h7', 'h13', 'h17', 'h23', 'h24', 'h22', 'h25'])]
melted_df2['Topics'].unique()
melted_df2
```

```python
melted_df2['Topics'] = melted_df2['Topics'].replace({'h0':'Contains themes of Action/Crime',
                             'h8': 'Contains themes of Action/Crime',
                            'h2':  'Contains themes of Adult Content and Explicit Language',
                            'h7':  'Contains themes of Family/Business Relations',
                            'h13': 'Contains themes of Period Romance', 
                            'h17': 'Contains themes of Royalty and Kingdom',
                            'h23': 'Contains themes of Royalty and Kingdom',
                            'h24': 'Contains themes of Royalty and Kingdom',
                            'h22': 'Contains themes of Lawyer/Media', 
                            'h25': 'Contains themes of Performance/Audience'})
melted_df2['Year'] = melted_df2['Year'].astype(int)
melted_df2['Topics'].unique()
```

```python
melted_df2.to_csv('melted.csv')
```
