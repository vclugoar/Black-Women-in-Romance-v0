```python
import pandas as pd 

df = pd.read_csv('Books2.csv')
```

```python
df2 = df[['index', 'Book', 'Author', 'Image', 'Year', 'Description', 'Buy',  'h0', 'h1', 
       'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'h8', 'h9', 'h10', 'h11', 'h12',
       'h13', 'h14', 'h15', 'h16', 'h17', 'h18', 'h19', 'h20', 'h21', 'h22',
       'h23', 'h24', 'h25', 'h26', 'h27', 'h28', 'h29']]

df2
```

```python
melted_df = df2.melt(id_vars=['index', 'Book', 'Author', 'Image', 'Year', 'Description', 'Buy'], 
        var_name= 'Topics', 
        value_name="Value")

melted_df = melted_df[melted_df['Value'] > 0]

melted_df[melted_df['Book'] == 'Stepping to a New Day']
```
